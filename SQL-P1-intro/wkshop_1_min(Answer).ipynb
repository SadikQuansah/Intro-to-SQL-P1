```python
#1.2
import sqlite3
import datetime

# Connect to the SQLite database
conn = sqlite3.connect('../data/mtcars.sqlite')
cursor = conn.cursor()

# Add a new column for vehicle age
alter_query = "ALTER TABLE results ADD COLUMN vehicle_age INTEGER"
cursor.execute(alter_query)

# Calculate vehicle age and update the column
current_year = datetime.datetime.now().year
update_query = "UPDATE results SET vehicle_age = ? - CAST(year AS INTEGER)"
cursor.execute(update_query, (current_year,))

# Commit the changes to the database
conn.commit()

# Close the database connection
conn.close()

```

```python
# 1.3

import sqlite3
import datetime

# Connect to the SQLite database
conn = sqlite3.connect('../data/mtcars.sqlite')
cursor = conn.cursor()

# Add a new column for the age label
alter_query = "ALTER TABLE results ADD COLUMN age_label VARCHAR(10)"
cursor.execute(alter_query)

# Update the age label column based on vehicle age
update_query = "UPDATE results SET age_label = CASE WHEN vehicle_age >= 45 THEN '45+' ELSE '<45' END"
cursor.execute(update_query)

# Commit the changes to the database
conn.commit()

# Close the database connection
conn.close()



```

```python
#2.2.7

import sqlite3
import pandas as pd

# Create a connection to the SQLite database
conn = sqlite3.connect('song.sqlite')

# Construct the SQL query with subqueries
query = """
SELECT 
    (SELECT AVG(song_popularity) FROM song_data WHERE LOWER(song_name) LIKE '%love%') AS avg_popularity_with_love,
    (SELECT AVG(song_popularity) FROM song_data WHERE LOWER(song_name) NOT LIKE '%love%') AS avg_popularity_without_love,
    (SELECT COUNT(*) FROM song_data WHERE LOWER(song_name) NOT LIKE '%love%') AS count_without_love
"""

# Execute the query and read the result into a DataFrame
pd.read_sql(query, con=conn)



```
